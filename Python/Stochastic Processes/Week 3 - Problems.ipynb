## Problema 1

Si tenemos el proceso $Y(t)=cos(\omega_1t)$, grafica su densidad espectral $S_Y(\omega)$ y su función de auto-correlación $\varphi_Y(t)$.


```python
#Primero, conseguimos los valores de nuestra función Y(t).
#Para esto, consideramos w=1. Tenemos:

w=2*np.pi #Es un vaor arbitrario
t=np.linspace(-2*np.pi,2*np.pi,1000)
Y=np.cos(w*t)

#Graficamos para observar su comportamiento.
plt.figure()
plt.plot(t,Y,'red',lw=3,label='$cos(\omega_1t)$')
plt.xlim(-2*np.pi,2*np.pi)
plt.legend()
```

```python
#Ahora, observaremos su función de autocorrelación.
#Para esto, convertimos Y(t) es una serie de Pandas.

auto_corr=pd.Series(Y)

#Observamos que si sean los datos.
auto_corr
```

```python
#Ahora, su funcion de autocorrelacion.
pd.plotting.autocorrelation_plot(auto_corr,lw=2,color='red')
```

```python
#Ahora, observemos su densidad espectral.
plt.figure()
plt.psd(Y,lw=2,color='red')
```

## Problema 2

Haz lo mismo que en el **Problema 1** pero esta vez con el proceso dinámico $Y(t)=\xi(t)$, donde $\xi(t)$ es *white noise*.

```python
#Primero, generamos el ruido blanco con una distribución normal.
#La distribucion la haremos a partir de numeros aleatorios como aprendimos en el curso.

mu=0    #La media.
sigma=1 #La desviación estandar.
N=10000 #El numero de muestras totales.

w_noise=np.random.normal(mu,sigma,N)
#Ahora, graficamos
x=np.linspace(0,N,N)
plt.figure()
plt.plot(w_noise,lw=0.5,color='r',label='White Noise')
plt.xlim(0,N)
plt.legend()
```

## Homework 3-1

Calculate the auto-correlation function $\varphi_Y(t)$ for a dynamic process $Y(t)=Asin(\omega_1t+\pi)$.

```python
#Primero obtenemos la función Y. Usaremos un rango de -2pi a 2pi y w=2pi.
t=np.linspace(-2*np.pi,2*np.pi,1000)
w=2*np.pi
A=2         #Valor arbitrario.
Y=A*np.sin(w*t+np.pi)

#Grafica la función.
plt.figure()
plt.plot(t,Y,color='red',lw=2,label='$Y(t)=Asin(\omega_1t+\pi)')
plt.xlim(-2*np.pi,2*np.pi)
plt.legend()
```

```python
#Ahora, observemos su auto-correlacion.
auto_corr=pd.Series(Y)
pd.plotting.autocorrelation_plot(auto_corr,lw=2,color='red')
```

```python
#Ahora, checaremos la grafica pero con la ecuación.
auto_corr_teo=((A**2)/4)*np.sin(w*t+np.pi)
x=np.linspace(0,1000,1000)

plt.figure()
plt.plot(x,auto_corr_teo,color='red',lw=2)
plt.xlim(0,1000)
```

## Homework 3-2

Calculate the auto-correlation function $\varphi_Y(t)$ for a dynamic process $Y(t)=Acos(\omega_1t)+B\xi(t)$, where $\xi(t)$ is the White noise.

```python
# Primero, tenemos que crear la función Y(t).
#Tenemos:
A=2        #Valor arbitrario.
B=3        #Valor arbitrario.
w=2*np.pi  #Valor arbitrario.
t=np.linspace(-2*np.pi,2*np.pi,1000)
P1=A*np.cos(w*t) #Siendo esta la parte cosenoidal de nuestra función.

#Ahora, esta es la parte del white noise.
mu=0    #La media.
sigma=1 #La desviación estandar.
N=1000 #El numero de muestras totales.

w_noise=np.random.normal(mu,sigma,N)  #Siendo este nuestro white noise.
Y=P1+B*w_noise #La funcion completa.

#Ahora, graficamos los resultados.
plt.figure()
plt.plot(t,Y,color='red',lw=2)
```

```python
#Ahora, checaremos resultados con la función de autocorrelación.
auto_corr=pd.Series(Y)
pd.plotting.autocorrelation_plot(auto_corr,lw=2,color='red')
```

## Homework 3-3

Estimate the diffusion constand $D$ of spherical particles with radius $a=1\mu m$ immersed in water at $T=300k$ using the Stokes-Einstein relation:

\begin{equation}
D=\frac{k_BT}{6\pi a\eta}
\end{equation}

and the following parameters

\begin{itemize}
\item Viscosity of water at room temperature: $\eta=0.85x10^{-3}Pa s$
\item The Boltzmann constant: $k_B=1.38064852x10^{-23}J K^{-1}$
\end{itemize}

```python
k=1.38064852*(10**(-23))
n=0.85*(10**(-3))
a=1*(10**(-6))
T=300


def diffusion(a,T,n,k):
    return (k*T)/(6*np.pi*a*n)

print('La difusión es',diffusion(a,T,n,k),'m^2/s')
```
